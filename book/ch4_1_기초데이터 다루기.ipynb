```python
#조건으로 데이터 선택하기

# boolean 연산을 통해 원하는 데이터만 추출

# 참고 - 다시 로딩#

flight = pd.read_csv('./Clean_Dataset.csv', encoding='cp949')
flight
```

```python
# 조건 : price가 12000 초과이고, airline이 Air_india인 항목만 추출


# boolean 연산으로 조건을 만족하는 데이터만 추출하기
flight_extract = flight[(flight['price']>12000) & (flight['airline']=='Air_India')]
flight_extract.head()
```

```python
# 조건이 너무 많거나 복잡해질수 있기 때문에
# 조건을 변수로 지정하여 사용할수도 있다

#조건을 별도로 저장하여 데이터 추출하기
#가격 : 12,000 초과
price_tag = flight['price']>12000

#항공사 이름
airline_tag = flight['airline']=='Air_India'

flight_extract_head = flight[price_tag & airline_tag].head()
flight_extract_head
```

```python
# 필요데이터 변경하기
```

```python
# 데이터 추가하기

#새로운 칼럼 만들기
flight['price2'] = flight['price']*2
flight.head()
```

```python
#기존 칼럼 연산으로 새로운 칼럼 만들기
flight['price3'] = flight['price'] + flight['price2']
flight.head()
```

```python
# insert 함수를 사용하여 원하는 위치에 칼럼을 생성할 수 있음

# df.insert(loc, column, value, allow_duplicates=False)

# 새로운 칼럼을 원하는 위치에 넣기
# 처음 0부터 시작하여 10번 칼럼으로 insert
flight.insert(10, 'duration2', flight['duration']*10)

flight.head()
```

```python
#데이터 삭제하기
#drop 메소드 활용
#axis 파라미터를 이용하여 행 기준 삭제인지, 칼럼 기준 삭제인지 정의할 수 있음
#axis = 0 열 레벨로 데이터를 삭제
#axis = 1 행 레벨로 데이터를 삭제
```

```python
#drop 메소드 사용하여 데이터 삭제하기 
flight.drop('price3', axis=1).head()
```

```python
#axis=0을 수행하면 행 레벨로 첫 번째 행이 삭제
flight.drop(index=0, axis=0).head()
```

```python
#원본 데이터프레임에서는 price3가 아직 삭제 되지 않았음.
#원본 데이터프레임에서 삭제하는 방법
#1. drop 메소드를 활용하여 데이터를 지운 후 다은 데이터프레임에 저장
#2. drop 메소드의 inplace 파라미터를 이용하여 원본 데이터에서 지우는 것(True)

#1
#flight['price3']=flight['price']+flight['price2']
flight=flight.drop('price3', axis=1)
flight.head()
```

```python
#2
#flight['price3'] = flight['price']+flight['price2']
flight.drop('price3', axis=1, inplace=True)
flight.head()
```

```python
#칼럼명 변경하기
#단순하거나 직관적인 칼럼명을 위해 rename 메소드 활용
```

```python
#rename 사용하여 칼럼명 변경
flight = flight.rename(columns={'airline':'airline_name', 'source_city':'departure_city'})
flight.head()

# rename에도 inplace 속성이 있기하네
```

```python
# 데이터프레임 정렬하기 ##
# sort_values 메소드 활용
```

```python
flight = flight.sort_values(by='Unnamed: 0', ascending=False)
flight.head()
```
