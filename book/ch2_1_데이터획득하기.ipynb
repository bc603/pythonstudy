1. Numpy로 데이터 가져오기

```python
#Numpy 가져오기
import numpy as np

#Numpy 데이터 만들기
a = np.array([1,2,3,4,5])
b = np.array(['대한민국', '포르투칼', '가나', '우루과이'])
c = np.array([1,2,'대한민국','포르투칼'])

#출력하기
print(a)
print(b)
print(c)
```

2. 판다스 이용하기

```python
#판다스 이용하기
import pandas as pd

#시리즈(Series) 생성하기
s = pd.Series(['대한민국', '포르투칼', '가나', '우루과이'], index=['가','나','다','라'], name='2022 카타르 월드컵')

#출력하기
print(s)
```

3. 데이터프레임 확인하기

```python
#데이터프레임 - Series가 여러개 합쳐진 자료형

#딕셔너리(Dictionary) 형으로 데이트프레임(Dataframe) 생성하기
a1 = pd.DataFrame({'a' : [1,2,3], 'b' : [4,5,6], 'c' : [7,8,9]})
a1

#key, value 형태
#{key1 : value1, key2 : value2, key3 : value3} 형태
#딕셔너리의 키값을 튜플은 사용할수 있지만, 리스트는 사용할 수 없다
```

```python
#리스트(List)형으로 데이터프레임(DataFrame) 생성하기
a2 = pd.DataFrame([[1,2,3], [4,5,6,], [7,8,9,]], ['a', 'b', 'c'], ['*', '&', '#'])
print(a2)

#튜플(Tuple) 자료형
#리스트는 []로 감싸지만, 튜플은 ()로 감싼다
#리스트는 요소의 수정, 추가, 삭제가 가능하지만, 튜플은 요소의 수정, 추가, 삭제가 불가능
```
